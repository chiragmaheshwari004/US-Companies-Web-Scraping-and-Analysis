# Web Scraping with Python and BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

headers = {
    "User-Agent": "Mozilla/5.0"
}


page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
df['Revenue (USD billions)'] = pd.to_numeric(df['Revenue (USD billions)'])

df['Employees'] = df['Employees'].str.replace(',', '')
df['Employees'] = pd.to_numeric(df['Employees'])
```

```python
df.info()
```

```python
top_revenue = df.sort_values(
    by='Revenue (USD billions)',
    ascending=False
)

top_revenue.head(10)
```

```python
import matplotlib.pyplot as plt

top_revenue.head(10).plot(
    x='Name',
    y='Revenue (USD billions)',
    kind='bar'
)

plt.title('Top 10 Companies by Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.savefig('Top_Revenue_Companies.png')
```

```python
top_employees = df.sort_values(
    by='Employees',
    ascending=False
)

top_employees.head(10)
```
